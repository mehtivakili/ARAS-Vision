```python
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output

# Set video paths
video_path = "/content/drive/MyDrive/deepsort_parameters/demo4.mp4"  # Change to "/content/demo2.avi" if needed
output_path = "/content/output_video.avi"

# Load VideoCapture
vid = cv2.VideoCapture(video_path)

# Get video properties
fps = int(vid.get(cv2.CAP_PROP_FPS))
frame_width = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define VideoWriter for saving output
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

while True:
    ret, frame = vid.read()
    if not ret:
        break

    # Preprocess frame
    frame_img, ratio = yolox_preprocess(frame, input_size=input_size)
    frame_tensor = torch.Tensor(frame_img).unsqueeze(0).to(device)

    # Detection with YOLOX
    with torch.no_grad():
        dets = model(frame_tensor)
    dets = postprocess(dets, 1, 0.5, 0.7, class_agnostic=True)[0]

    if dets is not None:
        # Rescale coordinates from letterbox back to the original frame size
        dets[:, 0] = (dets[:, 0]) / ratio
        dets[:, 1] = (dets[:, 1]) / ratio
        dets[:, 2] = (dets[:, 2]) / ratio
        dets[:, 3] = (dets[:, 3]) / ratio
        dets[:, 4] *= dets[:, 5]
        dets = dets[:, [0, 1, 2, 3, 4, 6]].cpu().numpy()
    else:
        dets = np.empty((0, 6))

    # Update tracker
    res = tracker.update(dets, frame)

    # Plot results
    tracker.plot_results(frame, show_trajectories=True)

    # Write processed frame to output video
    out.write(frame)

    # Display the frame in Colab
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    plt.figure(figsize=(8, 6))
    plt.imshow(frame)
    plt.axis('off')
    clear_output(wait=True)  # Clears previous output to update frames
    display(plt.gcf())  # Displays the current frame
    plt.close()

# Release resources
vid.release()
out.release()

print(f"✅ Processing complete! Output saved at: {output_path}")
```

```python
# Install required libraries
# !pip install gdown torch torchvision numpy matplotlib opencv-python boxmot
# !pip install numpy scipy scikit-learn

import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
from pathlib import Path
import gdown
from yolox.exp import get_exp
from yolox.utils import postprocess, fuse_model
from boxmot import BotSort
from boxmot.utils.ops import yolox_preprocess
import time
from collections import defaultdict, deque

# Dictionary for YOLOX model weights URLs
YOLOX_ZOO = {
    'yolox_n.pt': 'https://drive.google.com/uc?id=1AoN2AxzVwOLM0gJ15bcwqZUpFjlDV1dX',
    'yolox_s.pt': 'https://drive.google.com/uc?id=1uSmhXzyV1Zvb4TJJCzpsZOIcw7CCJLxj',
    'yolox_m.pt': 'https://drive.google.com/uc?id=11Zb0NN_Uu7JwUd9e6Nk8o2_EUfxWqsun',
    'yolox_l.pt': 'https://drive.google.com/uc?id=1XwfUuCBF4IgWBWK2H7oOhQgEj9Mrb3rz',
    'yolox_x.pt': 'https://drive.google.com/uc?id=1P4mY0Yyd3PPTybgZkjMYhFri88nTmJX5',
}

# Preprocessing pipeline
input_size = [640, 960]
cuda_available = torch.cuda.is_available() and torch.cuda.device_count() > 0
device = torch.device('cuda' if cuda_available else 'cpu')
print(f"CUDA available: {cuda_available}, Device count: {torch.cuda.device_count()}")
print(f"Running on device: {device}")
yolox_model = 'yolox_s.pt'
yolox_model_path = Path(yolox_model)

# Download model if not present
if not yolox_model_path.exists():
    gdown.download(YOLOX_ZOO[yolox_model], output=str(yolox_model_path), quiet=False)

# Initialize YOLOX model
exp = get_exp(None, 'yolox_s')
exp.num_classes = 1
ckpt = torch.load(yolox_model_path, map_location=device)
model = exp.get_model()
model.load_state_dict(ckpt["model"])
model = fuse_model(model).to(device).eval()

# Set video paths
video_path = "/content/drive/MyDrive/deepsort_parameters/demo3.mp4"
output_path = "/content/output_video.avi"

# Load VideoCapture
vid = cv2.VideoCapture(video_path)
if not vid.isOpened():
    raise ValueError("Could not open video file")

# Get video properties
fps = int(vid.get(cv2.CAP_PROP_FPS))
frame_width = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))
print(f"Video FPS: {fps}, Resolution: {frame_width}x{frame_height}")

# Define VideoWriter for saving output
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

# User-defined processing FPS
process_fps = float(input("Enter the desired processing FPS (e.g., 1, 2, 5): "))
if process_fps <= 0 or process_fps > fps:
    raise ValueError(f"Processing FPS must be between 0 and {fps} (video FPS)")
frames_per_process = int(fps / process_fps)
print(f"Processing every {frames_per_process}th frame to achieve {process_fps} FPS")

# User-defined max age in minutes for track_buffer
max_age_minutes = float(input("Enter the max age for tracks in minutes (e.g., 1, 2): "))
if max_age_minutes <= 0:
    raise ValueError("Max age must be greater than 0")
track_buffer = int(max_age_minutes * 60 * fps)
print(f"Track buffer set to {track_buffer} frames ({max_age_minutes} minutes at {fps} FPS)")

# Pixel-to-meter conversion factor (adjust this based on your video)
pixels_per_meter = 100  # Assumption: 100 pixels = 1 meter; adjust after calibration
print(f"Using {pixels_per_meter} pixels per meter for distance calculation")

# Initialize tracker with updated track_buffer
tracker = BotSort(
    reid_weights=Path('osnet_x0_25_msmt17.pt'),
    device=device,
    half=False,
    track_high_thresh=0.5,
    track_low_thresh=0.1,
    new_track_thresh=0.6,
    track_buffer=track_buffer,
    match_thresh=0.8
)

# Process video at user-defined FPS
frame_count = 0
processed_frame_count = 0
max_frames = 4000
start_time = time.time()
plt.ion()

# Store the last processed frame for copying annotations to skipped frames
last_annotated_frame = None

# Track positions of IDs over the last 5 processed steps
id_positions = defaultdict(lambda: deque(maxlen=5))  # {id: deque([(center_x, center_y), ...])}
crop_counters = defaultdict(int)  # {id: counter} for naming files like ID2_1.png, ID2_2.png, ...

while frame_count < max_frames:
    ret, frame = vid.read()
    if not ret:
        break

    # Process only at the specified FPS
    if frame_count % frames_per_process == 0:
        # Preprocess frame
        frame_img, ratio = yolox_preprocess(frame, input_size=input_size)
        frame_tensor = torch.Tensor(frame_img).unsqueeze(0).to(device)

        # Detection with YOLOX
        with torch.no_grad():
            dets = model(frame_tensor)
        dets = postprocess(dets, 1, 0.5, 0.7, class_agnostic=True)[0]

        if dets is not None:
            # Rescale coordinates
            dets[:, 0] = dets[:, 0] / ratio
            dets[:, 1] = dets[:, 1] / ratio
            dets[:, 2] = dets[:, 2] / ratio
            dets[:, 3] = dets[:, 3] / ratio
            dets[:, 4] *= dets[:, 5]
            dets = dets[:, [0, 1, 2, 3, 4, 6]].cpu().numpy()
            print(f"Detections: {len(dets)}")
        else:
            dets = np.empty((0, 6))
            print("No detections")

        # Update tracker
        res = tracker.update(dets, frame)

        # Store positions of current IDs
        current_positions = {}
        for track in res:
            x1, y1, x2, y2, track_id, conf, cls, *_ = track
            x1, y1, x2, y2, track_id = map(int, (x1, y1, x2, y2, track_id))
            # Calculate center of bounding box
            center_x = (x1 + x2) / 2
            center_y = (y1 + y2) / 2
            current_positions[track_id] = (x1, y1, x2, y2)  # Store bbox for cropping
            id_positions[track_id].append((center_x, center_y))

        # Check distances between pairs of IDs over the last 5 steps
        if processed_frame_count >= 4:  # Need at least 5 processed frames
            active_ids = list(id_positions.keys())
            for i in range(len(active_ids)):
                for j in range(i + 1, len(active_ids)):
                    id1, id2 = active_ids[i], active_ids[j]
                    # Check if both IDs have 5 steps of history
                    if len(id_positions[id1]) == 5 and len(id_positions[id2]) == 5:
                        # Calculate distances over the 5 steps in pixels
                        distances_pixels = []
                        for step in range(5):
                            pos1 = id_positions[id1][step]
                            pos2 = id_positions[id2][step]
                            dist = np.sqrt((pos1[0] - pos2[0])**2 + (pos1[1] - pos2[1])**2)
                            distances_pixels.append(dist)
                        # Convert distances to meters
                        distances_meters = [dist / pixels_per_meter for dist in distances_pixels]
                        # Check if distance is decreasing over the 5 steps
                        if all(distances_meters[k] > distances_meters[k + 1] for k in range(len(distances_meters) - 1)):
                            # Check if the total decrease is at least 1 meter
                            distance_decrease = distances_meters[0] - distances_meters[-1]
                            if distance_decrease >= 1.0:
                                print(f"IDs {id1} and {id2} are getting closer over 5 steps by {distance_decrease:.2f}m: {distances_meters}")
                                # Ensure both IDs are still in the current frame
                                if id1 in current_positions and id2 in current_positions:
                                    # Crop the frame around the two IDs
                                    bbox1 = current_positions[id1]
                                    bbox2 = current_positions[id2]
                                    # Compute the combined bounding box with padding
                                    padding = 50  # Add padding around the combined bbox
                                    x_min = max(0, min(bbox1[0], bbox2[0]) - padding)
                                    y_min = max(0, min(bbox1[1], bbox2[1]) - padding)
                                    x_max = min(frame_width, max(bbox1[2], bbox2[2]) + padding)
                                    y_max = min(frame_height, max(bbox1[3], bbox2[3]) + padding)
                                    # Crop the frame
                                    cropped_frame = frame[int(y_min):int(y_max), int(x_min):int(x_max)]
                                    # Save as PNG with ID-based naming
                                    base_id = min(id1, id2)  # Use the smaller ID for naming
                                    crop_counters[base_id] += 1
                                    crop_filename = f"/content/ID{base_id}_{crop_counters[base_id]}.png"
                                    cv2.imwrite(crop_filename, cropped_frame)
                                    print(f"Saved cropped frame as {crop_filename}")
                                else:
                                    print(f"Skipping crop: One or both IDs ({id1}, {id2}) not in current frame")

        # Plot results with trajectories
        tracker.plot_results(frame, show_trajectories=True)

        # Manually add larger ID labels
        for track in res:
            x1, y1, x2, y2, track_id, conf, cls, *_ = track
            x1, y1, x2, y2, track_id = map(int, (x1, y1, x2, y2, track_id))
            label = f"ID: {track_id}"
            cv2.putText(frame, label, (x1, y1 - 20), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 255, 0), 3)

        # Store the annotated frame
        last_annotated_frame = frame.copy()
        processed_frame_count += 1
        print(f"Processed frame {processed_frame_count} at video time {frame_count/fps:.1f}s")
    else:
        # For skipped frames, copy the last annotated frame's annotations
        if last_annotated_frame is not None:
            frame = last_annotated_frame.copy()
        else:
            # If no frame has been processed yet, use the current frame without annotations
            pass

    # Write frame to output video (all frames for smooth video)
    out.write(frame)

    # Display the frame (every processed frame for simplicity)
    if frame_count % frames_per_process == 0:
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        plt.figure(figsize=(10, 6))
        plt.imshow(frame_rgb)
        plt.title(f"Frame {frame_count} - Time: {frame_count/fps:.1f}s")
        plt.axis('off')
        clear_output(wait=True)
        display(plt.gcf())
        plt.close()

    frame_count += 1
    elapsed_time = time.time() - start_time
    # print(f"Progress: {frame_count}/{max_frames} frames ({frame_count/max_frames*100:.1f}%)")
    # print(f"Elapsed time: {elapsed_time:.1f}s")

# Release resources
vid.release()
out.release()
print(f"✅ Processing complete! Output saved at: {output_path}")
print(f"Total time: {time.time() - start_time:.1f}s")
```
